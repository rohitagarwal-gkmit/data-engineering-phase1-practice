# Pandas Introduction

Pandas is an open-source data manipulation and analysis library for the Python programming language. It provides easy-to-use data structures and data analysis tools for working with structured data, such as tabular data (like spreadsheets or SQL tables). The name "pandas" is derived from the term "panel data," which is a type of multi-dimensional data set commonly used in statistics and econometrics.

Pandas is particularly well-suited for tasks such as data cleaning, data transformation, and data analysis. It offers two primary data structures:

1. **Series**: A one-dimensional data structure that is essentially a labeled array, similar to a single column or row of data in a DataFrame.

2. **DataFrame**: A two-dimensional, tabular data structure resembling a spreadsheet with rows and columns. Each column can have a different data type, such as integers, floating-point numbers, strings, or dates.

Pandas provides a wide range of functions and methods for data manipulation and analysis, including:
- Data cleaning: handling missing data, data imputation, and data alignment
- Data filtering and selection
- Aggregation and summarization of data
- Data merging and joining
- Time series data manipulation
- Reading and writing data from/to various file formats, such as CSV, Excel, SQL databases, and more

The clip method is used to limit the values in a Series to a specified range.

## drop_duplicates
The drop_duplicates method is used to remove duplicate values from a Series.

```python
temp = pd.Series([1, 1, 2, 2, 3, 3, 4, 4])
print("Original:")
print(temp)
print(f"\nDuplicates count: {temp.duplicated().sum()}")
print("\nAfter drop_duplicates:")
print(temp.drop_duplicates())
```

## isnull and dropna
The isnull method is used to check for missing or NaN values. The dropna method removes missing values.

```python
temp = pd.Series([1, 2, 3, np.nan, 5, 6, np.nan, 8, np.nan, 10])
print("Original:")
print(temp)
print(f"\nNull count: {temp.isnull().sum()}")
```

```python
# dropna
print("After dropna:")
print(temp.dropna())
```

## fillna
The fillna method is used to fill missing values in a Series with specified values.

```python
# fillna with mean
print(f"Mean: {temp.mean()}")
print("\nAfter fillna with mean:")
print(temp.fillna(temp.mean()))
```

## isin
The isin method is used to filter a Series to include only elements that match a list of values.

```python
# isin
vk[vk.isin([49, 99, 100])]
```

## apply
The apply method is used to apply a function to each element of a Series.

```python
# Create sample movies data
movies = pd.Series(['Shah Rukh Khan', 'Amitabh Bachchan', 'Akshay Kumar', 'Salman Khan'], 
                   index=['Movie1', 'Movie2', 'Movie3', 'Movie4'], name='lead')
print("Original:")
print(movies)
print("\nAfter apply (first name uppercase):")
print(movies.apply(lambda x: x.split()[0].upper()))
```

## copy
The copy method is used to create a copy of a Series.

```python
# copy
new = vk.head().copy()
new.iloc[0] = 100
print("Modified copy:")
print(new)
print("\nOriginal (unchanged):")
print(vk.head())
```

```python
def euclidean(row):
    pt_A = row['1st point']
    pt_B = row['2nd point']
    return ((pt_A[0] - pt_B[0])**2 + (pt_A[1] - pt_B[1])**2)**0.5

points_df['distance'] = points_df.apply(euclidean, axis=1)
points_df
```

# Day 35 - GroupBy Object in Pandas (Part 1)

A GroupBy object is a crucial part of the data manipulation process for data aggregation and transformation.

```python
import pandas as pd
import numpy as np

# Create a sample DataFrame
data = {
    'Category': ['A', 'B', 'A', 'B', 'A'],
    'Value': [10, 20, 15, 25, 30]
}

df = pd.DataFrame(data)

# Group the data by the 'Category' column
grouped = df.groupby('Category')

# Calculate the sum of 'Value' for each group
sum_values = grouped['Value'].sum()
print(sum_values)
```

## Practical Use with Sample Data

```python
# Create sample movies data
np.random.seed(42)
movies = pd.DataFrame({
    'Series_Title': [f'Movie_{i}' for i in range(100)],
    'Genre': np.random.choice(['Action', 'Drama', 'Comedy', 'Horror', 'Sci-Fi'], 100),
    'Runtime': np.random.randint(90, 180, 100),
    'IMDB_Rating': np.random.uniform(6.0, 9.5, 100).round(1),
    'No_of_Votes': np.random.randint(10000, 1000000, 100),
    'Gross': np.random.randint(1000000, 500000000, 100),
    'Director': np.random.choice(['Director_A', 'Director_B', 'Director_C', 'Director_D'], 100),
    'Star1': np.random.choice(['Actor_A', 'Actor_B', 'Actor_C', 'Actor_D', 'Actor_E'], 100)
})

movies.head()
```

```python
# Group by Genre
genres = movies.groupby('Genre')

# Sum aggregation
genres[['Runtime', 'Gross']].sum()
```

```python
# Mean aggregation
genres[['Runtime', 'IMDB_Rating', 'No_of_Votes']].mean()
```

```python
# Find top 3 genres by total earnings
movies.groupby('Genre')['Gross'].sum().sort_values(ascending=False).head(3)
```

```python
# Find genre with highest avg IMDB rating
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

```python
# Find director with most popularity (votes)
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

```python
# Find number of movies done by each actor
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False)
```

# Day 37 - Joins in Pandas

Joins in pandas refer to the process of combining data from two or more DataFrames based on a common column or index.

```python
import pandas as pd
import numpy as np
```

## Types of Joins

1. **Inner Join**: Returns only matching rows
2. **Left Join**: Returns all rows from left DataFrame
3. **Right Join**: Returns all rows from right DataFrame
4. **Full Outer Join**: Returns all rows from both DataFrames

```python
# Create sample data
students = pd.DataFrame({
    'student_id': [1, 2, 3, 4, 5],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'partner': [2, 1, 4, 3, 5]
})

courses = pd.DataFrame({
    'course_id': [1, 2, 3, 4, 5],
    'course_name': ['Python', 'SQL', 'ML', 'Data Science', 'Statistics'],
    'price': [2499, 1999, 3999, 4999, 2999]
})

enrollments = pd.DataFrame({
    'student_id': [1, 1, 2, 3, 3, 4],
    'course_id': [1, 2, 1, 3, 4, 5]
})

print("Students:")
print(students)
print("\nCourses:")
print(courses)
print("\nEnrollments:")
print(enrollments)
```

## Concat

```python
# Concat along axis 0 (rows)
pd.concat([enrollments, enrollments], ignore_index=True)
```

## Inner Join

```python
inner = students.merge(enrollments, how='inner', on='student_id')
inner
```

## Left Join

```python
left = courses.merge(enrollments, how='left', on='course_id')
left
```

## Right Join

```python
right = enrollments.merge(students, how='right', on='student_id')
right
```

## Outer Join

```python
outer = students.merge(enrollments, how='outer', on='student_id')
outer
```

# Day 40 - Stacking and Unstacking

Operations to reshape data between wide and long formats.

```python
import numpy as np
import pandas as pd

# Create a DataFrame with multi-level columns
df = pd.DataFrame(np.random.rand(4, 4), 
                  columns=[['A', 'A', 'B', 'B'], ['X', 'Y', 'X', 'Y']])
print("Original DataFrame:")
print(df)
```

## Stacking
Pivoting the innermost level of column labels to become the innermost level of row labels.

```python
# Stack the innermost level of columns
stacked_df = df.stack()
print("Stacked DataFrame:")
print(stacked_df)
```

## Unstacking
Pivoting the innermost level of row labels to become the innermost level of column labels.

```python
# Unstack the innermost level of row labels
unstacked_df = stacked_df.unstack()
print("Unstacked DataFrame:")
print(unstacked_df)
```

```python
# Using multiindex dataframe
print("\nOriginal branch_df1:")
print(branch_df1)
print("\nAfter unstack():")
print(branch_df1.unstack())
```

# Day 42 - Long vs Wide Data Formats

"Long" and "wide" describe two different ways of organizing and structuring data.

```python
import numpy as np
import pandas as pd
```

## Data Conversion: melt
Wide data is converted into long format using `melt`.

```python
# Simple melt example
print("Wide format:")
print(pd.DataFrame({'cse': [120], 'ece': [100], 'mech': [50]}))

print("\nLong format after melt:")
print(pd.DataFrame({'cse': [120], 'ece': [100], 'mech': [50]}).melt())
```

```python
# Using var_name and value_name
pd.DataFrame({'cse': [120], 'ece': [100], 'mech': [50]}).melt(
    var_name='branch', value_name='num_students')
```

```python
# Branch with year data
df_wide = pd.DataFrame({
    'branch': ['cse', 'ece', 'mech'],
    '2020': [100, 150, 60],
    '2021': [120, 130, 80],
    '2022': [150, 140, 70]
})

print("Wide format:")
print(df_wide)
```

```python
# Using id_vars to keep certain columns
df_long = df_wide.melt(id_vars=['branch'], var_name='year', value_name='students')
print("Long format:")
print(df_long)
```

# Day 43 - Pivot Table

Pivot tables allow you to summarize and aggregate data.

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Sample data
data = {
    'Date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02'],
    'Product': ['A', 'B', 'A', 'B'],
    'Sales': [100, 200, 150, 250],
}

df = pd.DataFrame(data)

# Create a pivot table
pivot_table = pd.pivot_table(df, values='Sales', index='Date', columns='Product', aggfunc='sum')
print(pivot_table)
```

# Day 45 - Date and Time in Pandas

```python
import numpy as np
import pandas as pd
```

## 1. Timestamp
Represents a single timestamp.

```python
# Creating Timestamp objects
print(pd.Timestamp('2023/1/5'))
print(pd.Timestamp('2023-1-5'))
print(pd.Timestamp('5th January 2023'))
print(pd.Timestamp('5th January 2023 9:21AM'))
```

```python
# Scatter plot with annotations
sample_df = batters.head(10)

plt.figure(figsize=(12, 8))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs']/20)

for i in range(len(sample_df)):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], 
             sample_df['batter'].values[i], fontsize=8)

plt.xlabel('Average')
plt.ylabel('Strike Rate')
plt.title('Batsmen Analysis')
plt.show()
```

## 3D Surface Plot

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
xx, yy = np.meshgrid(x, y)
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12, 8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx, yy, z, cmap='viridis')
fig.colorbar(p)
plt.title('3D Surface Plot: sin(x) + cos(y)')
plt.show()
```

## Contour Plots

```python
fig = plt.figure(figsize=(12, 8))
ax = plt.subplot()
p = ax.contourf(xx, yy, z, cmap='viridis', levels=20)
fig.colorbar(p)
plt.title('Filled Contour Plot')
plt.show()
```
